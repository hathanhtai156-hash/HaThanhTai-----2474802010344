```
# Bài 02: Liệt kê danh sách stopwords của các ngôn ngữ trong NLTK

import nltk
from nltk.corpus import stopwords

# Tải stopwords (chỉ cần chạy một lần)
nltk.download('stopwords')

# Liệt kê các ngôn ngữ có stopwords
languages = stopwords.fileids()

print("Các ngôn ngữ có trong stopwords của NLTK:")
for i, lang in enumerate(languages, start=1):
    print(f"{i}. {lang}")

# Hiển thị 10 stopwords đầu tiên của mỗi ngôn ngữ
print("\n==============================")
print("10 stopwords đầu tiên của mỗi ngôn ngữ")
print("==============================")

for lang in languages:
    print(f"\nNgôn ngữ: {lang}")
    print(stopwords.words(lang)[:10])
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download dữ liệu
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

text = "This is a simple example to show how to remove stopwords from a sentence."

stop_words = set(stopwords.words("english"))

words = word_tokenize(text)

filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("Văn bản ban đầu:")
print(text)

print("\nSau khi loại bỏ stopwords:")
print(filtered_words)
```

```
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

word = "good"

synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print("Từ cần tìm:", word)

print("\nTừ đồng nghĩa:")
print(set(synonyms))

print("\nTừ trái nghĩa:")
print(set(antonyms))
```

```
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

word = "dog"

synsets = wordnet.synsets(word)

syn = synsets[0]

print("Từ:", word)
print("Synset:", syn.name())
print("Định nghĩa:", syn.definition())

print("\nHypernyms - từ khái quát hơn:")
for h in syn.hypernyms():
    print(h.name(), ":", h.definition())

print("\nHyponyms - từ cụ thể hơn:")
for h in syn.hyponyms()[:10]:
    print(h.name(), ":", h.definition())
```

```
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

word = "happy"

synsets = wordnet.synsets(word)

print("Từ:", word)
print("Số lượng synset:", len(synsets))

for i, syn in enumerate(synsets, start=1):
    print("\nSynset", i)
    print("Tên:", syn.name())
    print("Loại từ:", syn.pos())
    print("Định nghĩa:", syn.definition())
    print("Ví dụ:", syn.examples())
    print("Lemma:", syn.lemma_names())
```

```
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

verb1 = "run"
verb2 = "walk"

syn1 = wordnet.synsets(verb1, pos=wordnet.VERB)[0]
syn2 = wordnet.synsets(verb2, pos=wordnet.VERB)[0]

similarity = syn1.path_similarity(syn2)

print("Động từ 1:", verb1, "-", syn1.name())
print("Động từ 2:", verb2, "-", syn2.name())
print("Độ tương đồng:", similarity)
```

```
import nltk
from nltk.corpus import names

# Tải dữ liệu
nltk.download('names')

# Đọc danh sách tên
male_names = names.words('male.txt')
female_names = names.words('female.txt')

# Hiển thị số lượng
print("Số lượng tên nam:", len(male_names))
print("Số lượng tên nữ:", len(female_names))

# Hiển thị 10 tên đầu tiên
print("\n10 tên nam đầu tiên:")
print(male_names[:10])

print("\n10 tên nữ đầu tiên:")
print(female_names[:10])

# Tìm tên xuất hiện ở cả nam và nữ
common_names = set(male_names).intersection(set(female_names))

print("\nSố tên xuất hiện ở cả nam và nữ:", len(common_names))

print("\n20 tên trùng nhau đầu tiên:")
print(sorted(common_names)[:20])
```
